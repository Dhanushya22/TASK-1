```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
df = sns.load_dataset("diamonds")
print("First 5 rows of the dataset:")
display(df.head())
print("\nDataset Info:")
df.info()
print("\nSummary Statistics:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())
df.drop_duplicates(inplace=True)
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribution of Diamond Prices")
plt.xlabel("Price ($)")
plt.show()
sns.boxplot(x="cut", y="price", data=df)
plt.title("Diamond Price by Cut")
plt.show()
sns.boxplot(x="color", y="price", data=df)
plt.title("Diamond Price by Color")
plt.show()
sns.scatterplot(x="carat", y="price", hue="cut", data=df)
plt.title("Diamond Price vs Carat Weight")
plt.show()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numeric Features")
plt.show()
sns.countplot(x="cut", data=df)
plt.title("Count of Diamonds by Cut")
plt.show()
print("\nKey Insights:")
print("- Price is strongly correlated with carat weight.")
print("- 'Ideal' cut diamonds are the most common.")
print("- Color and clarity also influence price, but in varying ways.")
print("- No missing values and minimal cleaning needed.")

```
